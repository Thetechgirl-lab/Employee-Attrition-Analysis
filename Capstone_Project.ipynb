# 3. FEATURE ENGINEERING


*   Attrition Encoding
*   Overtime Encoding


*   Tenure Group (Years at Company)
*   Income level buckets

The key engineered features include Attrition encoding, Overtime encoding, tenure grouping based on years at the company, and income level categorization. These features were selected due to their strong relevance to employee turnover and their interpretability in an HR analytics context.







```python
# Attrition Encoding
df['Attrition_encoded'] = df['Attrition'].map({'No': 0, 'Yes': 1})
```

```python
# Overtime Encoding
df['Overtime_encoded'] = df['OverTime'].map({'No': 0, 'Yes': 1})
```

```python
# Tenure Group (Years at company)
df['TenureGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins=[0, 2, 5, 10, 40],
    labels=['0–2 years', '3–5 years', '6–10 years', '10+ years']
)
```

```python
# Income level buckets
df['IncomeLevel'] = pd.qcut(
    df['MonthlyIncome'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

# 4. Correlation & Relationship Analysis

To identify key drivers of attrition, the following relationships were analyzed:

1. **Overtime** – Employees working overtime are more likely to leave.
2. **Monthly Income** – Lower income employees show higher attrition.
3. **Tenure** – Early-tenure employees are at higher risk of leaving.
4. **Work-Life Balance** – Poor work-life balance is strongly associated with attrition.
5. **Job Role / Job Level** – Certain roles have higher turnover risk.




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Attrition vs Overtime
overtime_crosstab = pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100
print("Attrition vs Overtime (%)\n", overtime_crosstab.round(2))

# 2️⃣ Attrition vs Monthly Income (Boxplot)
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Attrition')
plt.ylabel('Monthly Income')
plt.show()

# 3️⃣ Attrition vs Tenure Group
tenure_crosstab = pd.crosstab(df['TenureGroup'], df['Attrition'], normalize='index') * 100
print("\nAttrition vs Tenure Group (%)\n", tenure_crosstab.round(2))

# 4️⃣ Attrition vs Work-Life Balance
wlb_crosstab = pd.crosstab(df['WorkLifeBalance'], df['Attrition'], normalize='index') * 100
print("\nAttrition vs Work-Life Balance (%)\n", wlb_crosstab.round(2))

# 5️⃣ Attrition vs Job Role
jobrole_crosstab = pd.crosstab(df['JobRole'], df['Attrition'], normalize='index') * 100
print("\nAttrition vs Job Role (%)\n", jobrole_crosstab.round(2))

# 6️⃣ Numeric Correlation Heatmap (Optional but useful)
numeric_cols = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'TotalWorkingYears', 'Attrition_encoded']
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
```

# 5. PREDICTIVE MODEL

*   Select feature for the model

*   Split the data

*   Train logistics regression
*   Model Evaluation


*   Feature importance






```python
# Select feature for the model
features = ['Overtime_encoded', 'MonthlyIncome', 'YearsAtCompany', 'WorkLifeBalance']
target = 'Attrition_encoded'
```

```python
# Split the data
from sklearn.model_selection import train_test_split

X = df[features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# Train logistics regression
from sklearn.linear_model import LogisticRegression

# Train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Display coefficients and intercept (text output)
coefficients = pd.DataFrame({
    'Feature': features,
    'Coefficient': model.coef_[0]
})
print("Feature Coefficients:\n", coefficients)
print("\nIntercept:", model.intercept_)
```

```python
# Model evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Feature importance
importance = pd.DataFrame({
    'Feature': features,
    'Coefficient': model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)

print(importance)
```

## Predictive Modeling: Logistic Regression

A logistic regression model was trained to predict employee attrition using key features identified during exploratory data analysis and feature engineering. The features included:

- **OverTime**: Whether an employee works overtime.
- **Monthly Income**: The employee's monthly salary.
- **Years at Company**: Total tenure at the organization.
- **Work-Life Balance**: Self-reported work-life balance rating.

### Key Findings

- **OverTime** has the strongest positive impact on attrition. Employees who work overtime are significantly more likely to leave.
- **Monthly Income** is negatively associated with attrition. Lower-paid employees have a higher likelihood of leaving.
- **Years at Company** shows early-tenure employees (0–2 years) are at higher risk of leaving.
- **Work-Life Balance** confirms EDA results: poor work-life balance increases the probability of attrition.

### Model Performance

- Accuracy on the test set: 0.84%
- The confusion matrix and classification report indicate the model is reasonably effective at distinguishing between employees who leave and those who stay.



# 6. Insights / Recommendations

The model confirms the EDA observations and highlights actionable areas for intervention:
- Reducing excessive overtime could improve retention.
- Addressing salary gaps, especially for lower-income roles, may reduce turnover.
- Focusing retention strategies on early-tenure employees can help mitigate early attrition.
- Programs that enhance work-life balance may increase employee satisfaction and retention.
